```python
from twitterscraper import query_tweets
import datetime as dt
from time import sleep
```

```python

if __name__ == '__main__':
    for tweet in query_tweets('조국',1,lang='ko'):
        print(tweet.timestamp) #작성시간
        print(tweet.text)      #트윗내용
```

#### 날짜를 지정하여 출력이 가능하다.

```python
#fname = 'c:/Java/worldcup'+ str(dt.date.today())+'.txt' #저장이름.위치.날짜
for tweet in query_tweets('조국',1000,lang='ko',begindate=dt.date(2019,9,15), enddate=dt.date(2019,10,20)):
    print(tweet.timestamp)
    print(tweet.text)      #트윗내용
```
