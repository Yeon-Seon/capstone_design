```python
from pytrends.request import TrendReq
```

```python
import pandas as pd
keyword = ['조국']
pytrend = TrendReq(hl='ko', tz=540)
pytrend.build_payload(keyword, timeframe='today 5-y')
# Interest Over Time
cointrenddf = pytrend.interest_over_time()
```

```python
cointrenddf=cointrenddf.iloc[:,:1]
```

```python
print(cointrenddf.head())
```

```python
#관련 검색어가져오기.
pytrend.related_queries()['조국']['top']
```

```python
#연관검색어들
google_keyword = pytrend.related_queries()['조국']['top']['query']
#각 검색어 value값.
google_value = pytrend.related_queries()['조국']['top']['value']
```

```python
import numpy as np
import datetime
date_list=np.array(cointrenddf.index)
datestart=np.where(date_list ==np.datetime64(datetime.datetime(2019, 4, 28)))
print(datestart)
```

```python
cointrenddf=cointrenddf.iloc[234:,]
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.plot(cointrenddf,c='r', label="Google Trend")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
from wordcloud import WordCloud
#font_path = '/usr/share/fonts/truetype/nanum/NanumMyeongjoBold.ttf'
```

```python
font_path = 'C:/Windows/Fonts/BMDOHYEON_ttf.ttf'

wordcloud = WordCloud(
    font_path = font_path,
    width = 800,
    height = 800,
    background_color="white"
)
```

```python
keywords = {}

for keyword, value in zip(google_keyword, google_value):
     keywords[keyword.strip()] = value
        

wordcloud = wordcloud.generate_from_frequencies(keywords)        
```

```python
array = wordcloud.to_array()
print(type(array)) # numpy.ndarray
print(array.shape) # (800, 800, 3)


%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm

plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
